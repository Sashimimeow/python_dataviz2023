## ข้อ 4. Core Budget ไหน ('12-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายต่างๆ เกี่ยวกับน้ำท่วม',  '17-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายเพื่อการพัฒนากรุงเทพมหานคร', '28-เงินสำรองค่าใช้จ่าย ในการพัฒนาเขต') มีค่าเฉลี่ยงบประมาณที่ถูกอนุมัติสูงที่สุด

````
set(data1['Core_budget_code']) # ตรวจสอบผลลัพธ์ในคอลัมน์ Core_budget_code
````

````
# เอาค่า Null ออกก่อน
Dropnull = data1.fillna('-', inplace = False)
useful_data1 = Dropnull[Dropnull['Core_budget_code'] != '-']

# คำนวณค่าเฉลี่ยของแต่ละ Core budget
twelve = useful_data1[useful_data1['Core_budget_code'] == '12-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายต่างๆ เกี่ยวกับน้ำท่วม'][['Approved_buget']].mean()
seventeen = useful_data1[useful_data1['Core_budget_code'] == '17-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายเพื่อการพัฒนากรุงเทพมหานคร'][['Approved_buget']].mean()
twenty_eight = useful_data1[useful_data1['Core_budget_code'] == '28-เงินสำรองค่าใช้จ่าย ในการพัฒนาเขต'][['Approved_buget']].mean()

print(f'ค่าเฉลี่ยส่วนของเงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายต่างๆ เกี่ยวกับน้ำท่วม คือ {int(twelve):,} บาท')
print(f'ค่าเฉลี่ยส่วนของเงินงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายเพื่อการพัฒนากรุงเทพมหานคร คือ {int(seventeen):,} บาท')
print(f'ค่าเฉลี่ยส่วนของเงินสำรองค่าใช้จ่าย ในการพัฒนาเขต คือ {int(twenty_eight):,} บาท')

````

````
avg_core = int(twelve), int(seventeen), int(twenty_eight)
max(avg_core)

if max(avg_core) == int(twelve):
  print(f'ส่วนของเงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายต่างๆ เกี่ยวกับน้ำท่วม มีค่าเฉลี่ยงบประมาณที่ถูกอนุมัติสูงที่สุด ซึ่งมีค่าเท่ากับ {int(twelve):,} บาท')
else:
  if max(avg_core) == int(seventeen):
    print(f'ส่วนของเงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายเพื่อการพัฒนากรุงเทพมหานคร มีค่าเฉลี่ยงบประมาณที่ถูกอนุมัติสูงที่สุด ซึ่งมีค่าเท่ากับ {int(seventeen):,} บาท')
  elif max(avg_core) == int(twenty_eight):
    print(f'ส่วนของเงินสำรองค่าใช้จ่าย ในการพัฒนาเขต มีค่าเฉลี่ยงบประมาณที่ถูกอนุมัติสูงที่สุด ซึ่งมีค่าเท่ากับ {int(twenty_eight):,} บาท')
````
